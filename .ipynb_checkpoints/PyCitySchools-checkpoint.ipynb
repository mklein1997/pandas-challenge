```python
# Dependencies 
import pandas as pd
import numpy as np
```

```python
# Import data resources
schooldatacsv = "/Users/Michael/Documents/GT DA/pandas-challenge/Resources/schools_complete.csv"
studentdatacsv = "/Users/Michael/Documents/GT DA/pandas-challenge/Resources/students_complete.csv"
```

```python
# Read in csvs to pandas
schooldata = pd.read_csv(schooldatacsv)
studentdata = pd.read_csv(studentdatacsv)
```

```python
# Combine datasets
schooldatacombined = pd.merge(studentdata, schooldata, how="left", on=["school_name", "school_name"])
```

```python
schooldatacombined.columns
```

```python
# Calculate total number of schools and students
schooltotal = schooldatacombined['school_name'].nunique()
studenttotal = schooldatacombined['Student ID'].count()
print(schooltotal)
print(studenttotal)
```

```python
# Calculate the total budget
budgettotal = schooldatacombined['budget'].sum()
budgettotal
```

```python
# Calculate the average math and reading scores

mathaverage = schooldatacombined["math_score"].mean()
readingaverage = schooldatacombined["reading_score"].mean()

print(mathaverage)

print(readingaverage)
```

```python
# Calculate the percentage of students with a passing math score
passingmathstudents = schooldatacombined.loc[(schooldatacombined["math_score"] >= 70)]
percentpassingmath = passingmathstudents.count()/39170
print(percentpassingmath)
```

```python
# Calculate the percentage of students with a passing reading score
passingreadingstudents = schooldatacombined.loc[(schooldatacombined["reading_score"] >= 70)]
percentpassingreading = passingreadingstudents.count()/39170
print(percentpassingreading)
```

```python
# Calculate the percentage of students who passed math and reading
passedboth = schooldatacombined.loc[(schooldatacombined['reading_score'] >= 70)] + schooldatacombined.loc[(schooldatacombined['math_score'] >= 70)]
percentpassedboth = passedboth.count()/39170
percentpassedboth
```

```python
# Create a table breaking down school performance on spending ranges per student

schooloverviewdf["Per Student Budget ($)"] = schooloverviewdf["Per Student Budget ($)"].astype(str).astype(float)
schooloverviewdf["Average Math Score"] = schooloverviewdf["Average Math Score"].astype(str).astype(float)
schooloverviewdf["Average Reading Score"] = schooloverviewdf["Average Reading Score"].astype(str).astype(float)
schooloverviewdf["% Passing Math"] = schooloverviewdf["% Passing Math"].astype(str).astype(float)
schooloverviewdf["% Passing Reading"] = schooloverviewdf["% Passing Reading"].astype(str).astype(float)

bins = [0,584,629,644,675]
schooloverviewdf["Spending Ranges(Per Student)"] = pd.cut(schooloverviewdf["Per Student Budget ($)"], bins, include_lowest=True)
spendingps = schooloverviewdf.groupby("Spending Ranges(Per Student)")
spendingps[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading","% Passing Overall"]].mean()
```

```python
# Perform same operations based on school size

schooloverviewdf["Total Students"] = schooloverviewdf["Total Students"].astype(str).astype(float)

bins = [0,1000,2000,5000]
schooloverviewdf["School Size"] = pd.cut(schooloverviewdf["Total Students"], bins, include_lowest=True)
schoolsize = schooloverviewdf.groupby("School Size")
schoolsize[["Average Math Score","Average Reading Score", "% Passing Math", "% Passing Reading","% Passing Overall"]].mean()
```

```python
# Perform Same operations based on school type
schooloverviewdf["Per Student Budget ($)"] = schooloverviewdf["Per Student Budget ($)"].astype(str).astype(float)
schooloverviewdf["Average Math Score"] = schooloverviewdf["Average Math Score"].astype(str).astype(float)
schooloverviewdf["Average Reading Score"] = schooloverviewdf["Average Reading Score"].astype(str).astype(float)
schooloverviewdf["% Passing Math"] = schooloverviewdf["% Passing Math"].astype(str).astype(float)
schooloverviewdf["% Passing Reading"] = schooloverviewdf["% Passing Reading"].astype(str).astype(float)

schooltype = schooloverviewdf.groupby("School_Type")
schooltype[["Average Math Score","Average Reading Score", "% Passing Math", "% Passing Reading","% Passing Overall"]].mean()
    
    
```
